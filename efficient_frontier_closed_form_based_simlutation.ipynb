```python
# Author: Nopadon Juneam
# Date: 16 March 2021
# The source code is provided as extra material of Parallel Computing with CUDA Course.

import numpy as np
import pandas as pd

N=470

# load dataset from .csv file
stock_prices = pd.read_csv('./dataset/dataset1257x%s.csv' %N)
stock_prices.columns = ['Asset %s' % x for x in range(1, N+1)]
stock_prices
```

```python
# Compute the assets' returns
returns = (stock_prices/stock_prices.shift(1)).dropna()
returns
```

```python
#Compute the assets' expected returns
expected_returns = np.mean(returns)
expected_returns.to_numpy()
```

```python
#Compute the assets' return standard deviations
return_stds = np.std(returns)
return_stds.to_numpy()
```

```python
#Compute the assets' return covariance
return_covariances = np.cov(returns, rowvar=False)
return_covariances
```

```python
v_alpha = expected_returns.to_numpy()
m_zigma = return_covariances
v_unit = np.ones(N)

m_zigma_inv = np.linalg.inv(m_zigma)
a = np.dot(np.dot(v_alpha, m_zigma_inv), v_alpha)
b = np.dot(np.dot(v_alpha, m_zigma_inv), v_unit)
c = np.dot(np.dot(v_unit, m_zigma_inv), v_unit)

m_temp = np.zeros((2, 2))
m_temp[0][0] = a
m_temp[0][1] = b
m_temp[1][0] = b
m_temp[1][1] = c

m_temp_inv = np.linalg.inv(m_temp)
```

```python
def computeOptimalPortfolio(alpha_zero):
    m_a = np.ones(2)
    m_a[0] = alpha_zero
    v_lambda = np.dot(m_temp_inv, m_a)
    w_star = v_lambda[0]*np.dot(m_zigma_inv, v_alpha)+v_lambda[1]*np.dot(m_zigma_inv, v_unit)
    return w_star
```

```python
#Simulating efficient frontier with M samples
import matplotlib.pyplot as plt
M = 100
x = []
y = []
c = []
alpha_max = np.max(v_alpha)
alpha_min = np.min(v_alpha)
delta = (alpha_max-alpha_min)/M

for i in range(0, M+1):
    w_opt = computeOptimalPortfolio(alpha_min+(i*delta))
    y.append(np.dot(w_opt, v_alpha))
    x.append(np.dot(np.dot(w_opt, m_zigma), w_opt))
    c.append('red')

    rand = np.random.rand(N)
    portfolio = rand/np.sum(rand)
    y.append(np.dot(portfolio, expected_returns))
    x.append(np.dot(np.dot(portfolio, return_covariances), portfolio))
    c.append('green')

plt.scatter(x, y, color=c)
plt.show()
```
